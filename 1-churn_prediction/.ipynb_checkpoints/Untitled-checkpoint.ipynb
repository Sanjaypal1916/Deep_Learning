```python
import tensorflow as tf 
from tensorflow import keras
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("customer_churn.csv")
```

```python
df.head()
df.drop("customerID" , axis=1, inplace=True)
```

```python
def binaryconverter(text):
    map ={ "Yes": 1,
         "No": 0,
         "No internet service": 0, 
         "No phone service": 0}
    text = map[text]
    return text
```

```python
df.Dependents=df.Dependents.apply(binaryconverter)
df.Partner = df.Partner.apply(binaryconverter)
df.PhoneService= df.PhoneService.apply(binaryconverter)
df.MultipleLines = df.MultipleLines.apply(binaryconverter)
df.OnlineSecurity= df.OnlineSecurity.apply(binaryconverter)
df.OnlineBackup=df.OnlineBackup.apply(binaryconverter)
df.DeviceProtection=df.DeviceProtection.apply(binaryconverter)
df.TechSupport=df.TechSupport.apply(binaryconverter)
df.StreamingTV =df.StreamingTV.apply(binaryconverter)
df.StreamingMovies=df.StreamingMovies.apply(binaryconverter)
df.PaperlessBilling=df.PaperlessBilling.apply(binaryconverter)
df.Churn=df.Churn.apply(binaryconverter)
```

```python
df.drop("PaymentMethod", inplace=True, axis=1)
```

```python
# df["TotalCharges"] = df["TotalCharges"].apply(lambda x : float(x.replace(" ", ""))) #wrongmethod
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges,errors='coerce')
```

```python
df.TotalCharges = df.TotalCharges.dropna()
```

```python
df.shape
```

```python
df.TotalCharges.dtype
```

```python
map = {
    "Male":1,
    "Female":0
}
df.gender = df.gender.apply(lambda x : map[x])
```

```python

df1 = pd.get_dummies(data=df,  columns=["InternetService", "Contract"])
df1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
classs = ["tenure","MonthlyCharges", "TotalCharges"]
mms = MinMaxScaler()
df1[classs] = mms.fit_transform(df[classs])
```

```python
df1.tenure
```

```python
for col in df1:
    print(col + " : " + str(df1[col].unique()) +  " : " +str(df1[col].dtype))
```

```python

print("Any NaN in X_train?", np.isnan(df1).any())
```

```python
df1.TotalCharges.isnull().sum()
```

```python
df1.dropna(subset=["TotalCharges"], inplace=True)
```

```python
y = df1["Churn"]
x = df1.drop("Churn", axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
len(x.columns)
```

```python


model = keras.Sequential([
    keras.layers.Dense(20, input_shape=(22,), activation="relu"),
    keras.layers.Dense(10, activation="relu"),
    keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"]
             )


model.fit(X_train, y_train, epochs=100)
```

```python

print("Any NaN in X_train?", np.isnan(X_train).any())
print("Any Inf in X_train?", np.isinf(X_train).any())
print("Any NaN in y_train?", np.isnan(y_train).any())
print("Any Inf in y_train?", np.isinf(y_train).any())
```

```python
df1.TotalCharges.isnull().sum()
```

```python
df1.dropna(subset=["TotalCharges"], inplace=True)
```

```python
model.evaluate(X_test, y_test)
```

```python
test = model.predict(X_test)
y_pred=[]
for i in test:
    if i > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
y_pred
```

```python
from sklearn.metrics import confusion_matrix , classification_report
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
cr = classification_report( y_test,y_pred)
print(cr)
```
